```python
import json

with open('../dataset/originals/activity_net.v1-3_old.min.json', 'r') as f:
    old_dataset = json.load(f)
    
with open('../dataset/originals/activity_net.v1-3.min.json', 'r') as f:
    updated_dataset = json.load(f)
```

```python
assert old_dataset['version'] == updated_dataset['version']
for a,b in zip(old_dataset['taxonomy'], updated_dataset['taxonomy']):
    if a != b:
        print('Previous: {}'.format(a))
        print('Updated: {}'.format(b))
```

```python
for a in old_dataset['taxonomy']:
    if a['nodeId'] == 268:
        print('nodeId = 268', a)
    if a['parentId'] == 123:
        print('parentId = 123', a)
```

```python
old_keys = list(old_dataset['database'].keys())
updated_keys = list(updated_dataset['database'].keys())
old_keys.sort()
updated_keys.sort()
for k, j in zip(old_keys, updated_keys):
    if old_dataset['database'][k] != updated_dataset['database'][j]:
        print('Previous: {}\n{}'.format(k, old_dataset['database'][k]))
        print('Updated: {}\n{}'.format(j, updated_dataset['database'][j]))
        print('\n')
```
